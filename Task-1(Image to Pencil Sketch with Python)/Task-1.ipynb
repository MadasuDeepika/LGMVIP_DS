# Problem Statement: 

### <font color=red> We need to read the image in RBG format and then convert it to a grayscale image. This will turn an image into a classic black and white photo. Then the next thing to do is invert the grayscale image also called negative image, this will be our inverted grayscale image. Inversion can be used to enhance details. Then we can finally create the pencil sketch by mixing the grayscale image with the inverted blurry image. This can be done by dividing the grayscale image by the inverted blurry image. Since images are just arrays, we can easily do this programmatically using the divide function from the cv2 library in Python.

### Importing libraries 

```python
import cv2
import io # io Library used for various input and ouput operations
import cv2 # OpenCV
from PIL import Image
from IPython.display import display
import matplotlib.pyplot as plt # For interpreting data in graphs
%matplotlib inline
```

### Reading the image

```python
image = cv2.imread("tiger.jpg")
```

### Displaying the image

```python
plt.imshow(image)
```

### Reading the image in RBG format

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
```

### Converting the image to Grayscale image

```python
new_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
display(Image.fromarray(new_image))
```

### Inverting the image

```python
invert = cv2.bitwise_not(new_image) #Inverting the gray scale image into negative image.
display(Image.fromarray(invert))
```

### Adding Blur to the Inverse Grayscale image

```python
blur_image = cv2.GaussianBlur(invert, (5, 5), 0, 0) #It is is used for Removing Noise in the image
display(Image.fromarray(blur_image))
```

Then we can finally create the pencil sketch by mixing the grayscale image with the inverted blurry image. This can be done by dividing the grayscale image by the inverted blurry image.

### Now, we have our pencil sketch

```python
invert_blur_image = 255 - blur_image
pencil_sketch = cv2.divide(new_image, invert_blur_image, scale=250.0)
img_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
display(Image.fromarray(pencil_sketch))
```
